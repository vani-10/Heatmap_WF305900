Loading Data into pandas

```python
import pandas as pd
df = pd.read_excel("WF305900_matrix.xlsx")#We can read csv,txt,tsv files too
```

```python
print(df.head(3))
```

Reading data in pandas

```python
##Reading headers
print(df.columns)

##Read each column
print(df[['Peptide','Protein','Intensity_PP_S1701_078_scored.txt_0_0']][0:3])
#We can get single column or multiple columns with top 3 rows or more values

##Read each row
print(df.iloc[4])#To get first value of each of the column we use integer location[1] or iloc[6] to get 6th row value
print(df.iloc[0:6])#To get first 5 row values
#To get all the particular row in one go organised one below one
for index, row in df.iterrows():
    print(index,row["Peptide"])
#To get any desired value
df.loc[df["Protein"]=="1/ddb037180991"]

##Read a specific location (R,C)
print(df.iloc[2,1])#2nd row 1st column value
```

Sorting/Describing Data

```python
df.describe()#To get the glimse of statistics behind our data
```

```python
df.sort_values("Intensity_PP_S1701_078_scored.txt_0_0",ascending=False)#to sort in descending order
```

```python
df.sort_values(["Intensity_PP_S1701_078_scored.txt_0_0","Intensity_PP_S1701_040_scored.txt_0_63"],ascending=(1,0))
#This means 078 intensity is sorted in ascending order and 063 intensity in descending order
```
