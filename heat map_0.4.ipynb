Now the df has certain Nan values and we want to fill it with 0 so that it has no more nan values.There are 10 nan values which is found by using the df1.isnull().sum().sum() code

```python
import numpy as np
df1=np.nan_to_num(df)
```

The Intensities columns has values that vary in huge range so it has to be normalised between 0 to 1 and applying the below code normalization is done based on (x-xmin)/(xmax-xmin) formula

```python
from sklearn import preprocessing
d = preprocessing.normalize(df1)
scaled_df = pd.DataFrame(d)
scaled_df.head()
```

After normalization our Intensities value are in the range between 0 and 1 .When we plot the heatmap now we find that the colors are mostly in the 0.0 to 0.5 scale.So we will now calculate what is the count of the spread of the values by writing a snippet

```python
import seaborn as sns
heat_map = sns.heatmap(scaled_df)
```

```python
df_full=scaled_df
count=[0]*11
for i in range(64):
    for j in range(12655):
        df_full[i][j]=int((scaled_df[i][j])*10)
#print(df_full)
for i in range(64):
    for j in range(12655):
            count[int(df_full[i][j])]+=1
for i in range(11):
    print(count[i])
```

12655*65=809920 is the total intensities values out of which the range of 0 to 0.1 is 490041 followed by 0.1 to 0.2 are 265280 and so on.So it is very evident that the Intensities values are mostly spread between 0.0 to 0.4 and the values from 0.5 to 1.0 can be replaced by 0.4 by the snippet below and when we now generate a heatmap now.

```python
df_full=scaled_df
for i in range(64):
    for j in range(12655):
        if(df_full[i][j]>0.4):
            df_full[i][j]=0.4
```
